Importing the Dependencies


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

## **Importing the Boston house price dataset**

```python
house_price_dataset = sklearn.datasets.load_boston()
```

```python
from sklearn.datasets import fetch_openml
boston=fetch_openml(name='boston',version=1,as_frame=True)
df=boston.frame
df.head()
```

## **Features and Target**

```python
df['price']=boston.target
df.head()
```

## **Dividing the dataset into dependent and independent features**

```python
X = df.iloc[:,:-1]
X = X.astype('float')
X.head()
```

```python
y = df['price']
y.head()
y.astype('float')
```

## **Split the Data into Training & Test Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3, random_state=3)
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train.head(),X_test.head())
```

## **Linear Regression**

```python
lin_reg = LinearRegression()
```

```python
model=lin_reg.fit(X_train, Y_train)
```

```python
predicted=lin_reg.predict(X_test)
```

```python
score1 = metrics.r2_score(Y_test, predicted)
print("R square:",score1)
```

```python
score2 = metrics.mean_absolute_error(Y_test, predicted)
print("Mean Absolute Error:",score2)
```

```python
comparison_df = pd.DataFrame({'Actual Value': Y_test, 'Predicted Value': predicted})
print(comparison_df.head())
```
